## Code for Problem 1 - BFS

```python
import queue
def bfs(adj_list, start, end):
    root = {}
    queue.put(start)
    root[start] = -1
    visit[start] = 1
    while queue.qsize() > 0:
        start = queue.get()
        adj_list[start].sort()
        for i in range(len(adj_list[start])):
            current = adj_list[start][i]
            if visit[current]!=1:
                queue.put(current)
                visit[current] = 1
                root[current] = start
            if current == end:
                a = []
                val = end
                while val != -1:
                    a.insert(0, str(val+1))
                    val = root[val]
                print(len(a))
                print(" ".join(a))
                return
```

```python
cities, routes, transition, duration = map(int, input().split())
adj_list = []
visit = [0]*cities
queue = queue.Queue()
for i in range(cities):
    adj_list.append([])
while routes != 0:
    routes -= 1
    a, b = map(int, input().split())
    adj_list[a-1].append(b-1)
    adj_list[b-1].append(a-1)
start,end= map(int, input().split())
if start == end:
    print("1\n1")
else:
    bfs(adj_list, start-1, end-1)
```

## Code for Problem 2 - DFS

```python
from collections import defaultdict

import sys

def DFS(s, d, l, visit, level):
    visit[s] = True
    for node in d[s]:
        if visit[node] == False:
            if node in l.keys():
                l[node] = level+1
            DFS(node, d, l, visit, level+1)
    return l
```

```python
n = int(input())
d=defaultdict(list)
l={}
for i in range(n-1):
    u, v = map(int, input().split())
    d[u].append(v)
    d[v].append(u)
for i in range(int(input())):
    l[int(input())]=0
    visit=[0 for i in range(n+1)]
```

```python
l = DFS(1, d, l, visit, 0)
gf, dist= sys.maxsize, sys.maxsize
for key, value in l.items():
        if value < dist:
            dist = value
            gf = key
        if value == dist and key < gf:
            gf = key
print(gf)
```
