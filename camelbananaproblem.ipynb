## Aim
- To provide a programming solution to the Camel-Bananas Problem in which the parameters are the load capacity of the camel, the initial number of bananas and the distance needed to be covered between the start point and destination. 
- The Camel-Bananas Problem is as follows:
    - A person has X bananas and a camel. The camel cannot carry more than Y bananas at a time and eats a banana every km it travels. 
    - The person wants to transport the maximum number of bananas to a destination which is Z KMs away, using only the camel as a mode of transportation.
    - What is the maximum number of bananas that can be transferred to the destination using only camel (no other mode of transportation is allowed)?

## Code

```python
bananas = int(input("Total Number of Bananas"))
camel_cap = int(input("Maximum Number of Bananas the Camel can carry"))
distance = int(input("Distance between Start and Destination"))
```

With 3000 bananas:
- Camel takes 1000 to 1 km.
- 999 bananas remaining at 1 km but need one more to go back to start point so, at 1 km => 998. 
To transport the remaining bananas in the start point to the 1km point, the steps will have to be repeated twice except in the last iteration, the camel will not need to back to the original point. 

        - So at the 1 km point, the camel will have eaten a total of 5 bananas.

The above pattern will continue until the camel reaches the 200 km point. This is because a third trip to the beginning point will not be needed with only 2000 bananas remaining (200*5 = 1000 bananas eaten).

From the 200 km point there are only 2000 bananas:
- Camel takes 1000 to 201 km.
- 998 bananas left at 201 km (1 consumed to get to the point and another one taken in advance to get back to 200 km point)

    -      However to transport the remaining 1000 will only need 1 banana. 
    -      So from 200 km point, only 3 bananas are consumed for each 1 km traversed.

The above pattern will continue for another ~333 km until only 1000 bananas remain to be transported. 

At this point, we will only need to move forward so the remaining distance (1000 - 200 - 333) so 467 bananas will be eaten by the camel and the remaining 533 bananas can be sold in the market.

This is outlined in the code below.


```python
eaten = 0
remaining = bananas
for i in range(distance): #intermediate points are at every 1 km.
    while remaining > 0:
        remaining = remaining - camel_cap 
        if remaining == 1:
            eaten = eaten - 1
        eaten = eaten + 2
    eaten = eaten - 1
    remaining = bananas - eaten
    if remaining == 0:
        break
print(remaining)
```
